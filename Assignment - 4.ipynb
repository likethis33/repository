# 문제 11번
## 해결과정
- 해당 100개의 값을 문자열로 저장한다음 공백을 기준으로 나눠 리스트에 저장합니다.
- 100개의 리스트의 값들을 저장하는데 값의 크기가 너무 크므로 long형으로 저장했습니다.
- 모든 값이 더해진 후 그 길이를 출력합니다.
- 길이의 값에서 10을 빼면 42인데 10^42 만큼 total값을 나눠서 앞의 10자리 수를 얻습니다.

```python
max = 0

for x in range(1,1000000):
    cnt = 0
    n = x
    
    while(n != 1):
        if (n%2 == 0):
            n = n/2
        else:
            n = 3*n+1
        cnt += 1
        
    if (max<cnt):
        max = cnt
        max_n = x
        
print max_n
```
